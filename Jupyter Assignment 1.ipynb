```python
# Import statements

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats.stats import pearsonr 
import math
```

```python
# Initialize the data points to be plotted

year = list(range(1930,2020,10))
GDP = [1.015,1.33,2.29,3.26,4.951,6.759,9.366,13.131,15.599]
```

```python
# a) Plot using matplotlib

plt.plot(year, GDP, 'o--', color ='red')
plt.title('Real US GDP (in trillions)')
plt.suptitle('Figure 1: Real US GDP as a function of the year.')
plt.xlabel('Year')
plt.ylabel('Trillions of $')
plt.show()
```

```python
# b) Mathematical Relationship

# Correlation
print("The correlation coefficient bewtween the year and GDP is",np.corrcoef(year,GDP)[0][1])
```

1) The year and GDP have an exponential behavior and as the year increases the GDP increases exponentially. y = exp(x)

2) The correlation coefficient of the Year versus the GDP is 0.9642 which means that the Year and GDP are "positively correlated"

3) The above also displays a  "Monotonic graph"
